**مشخصات فردی**  
نام: صالح  
سن: ۳۶ سال  
مدرک تحصیلی: کارشناسی مهندسی نرم‌افزار  
دانشگاه: دانشگاه آزاد اسلامی  
معدل: ۱۷

**مهارت‌ها**  
- **برنامه‌نویسی:**  
  - مسلط به زبان‌های Python و C#  
  - تجربه کار با کتابخانه‌های یادگیری ماشین در پایتون  
  - تسلط بر تحلیل داده‌ها و الگوریتم‌های یادگیری ماشین  
  - استفاده از SQLAlchemy برای مدیریت ارتباط با پایگاه داده‌ها  
  - مدیریت پایگاه داده‌های SQL Server و اجرای کوئری‌های پیچیده  
  - استخراج داده‌ها و گزارش‌دهی از پایگاه داده به فایل‌های Excel  

- **پایگاه داده:**  
  - تسلط به MySQL  
  - استفاده از کتابخانه MySQL Connector برای ارتباط با پایگاه داده MySQL  
  - آشنایی با MongoDB  
  - بهینه‌سازی و مدیریت ارتباط با SQL Server  
  - تسلط بر مدیریت خطاها و استثناها در ارتباطات پایگاه داده  
  - توسعه پروژه‌های گزارش‌دهی و تحلیل داده با SQL Server  

- **یادگیری ماشین و تحلیل داده‌ها:**  
  - تحلیل عوامل موثر بر بیماری دیابت با استفاده از مدل‌های یادگیری ماشین  
  - توانایی تفسیر و تحلیل خروجی مدل‌های یادگیری ماشین  

- **کنترل نسخه و سیستم‌عامل‌ها:**  
  - توانایی کار با GitHub و مدیریت نسخه‌های کد  
  - آشنایی با لینوکس و کار با ابزارها و دستورات پایه سیستم‌عامل  
  - تسلط به سیستم‌عامل ویندوز و کار با نرم‌افزارهای مختلف تحت ویندوز  

- **نرم‌افزارها و محیط‌های توسعه:**  
  - تسلط به محیط Jupyter Notebook و VSCode  
  - کتابخانه‌های پایتون: Pandas، Scikit-learn، Numpy، OS، MySQL Connector  
  - Excel، SQL Server Management Studio (SSMS)  

**تجربه‌های حرفه‌ای**  
- **پروژه یادگیری ماشین:**  
  - تحلیل عوامل موثر بر بیماری دیابت با استفاده از داده‌های واقعی  
  - پیاده‌سازی مدل‌های پیش‌بینی و تفسیر نتایج با استفاده از پایتون و کتابخانه‌های یادگیری ماشین  
  - نوشتن مستندات پروژه و ارائه تحلیل‌های کامل به زبان فارسی و انگلیسی  

- **پروژه گزارش‌دهی SQL Server:**  
  - طراحی و توسعه سیستمی برای اجرای کوئری‌ها و استخراج داده‌ها از SQL Server  
  - بهینه‌سازی فرآیندها و توسعه خروجی‌های قابل فهم در قالب فایل‌های Excel  
  - افزودن امکانات جدید برای بهبود گزارش‌دهی و مدیریت داده  

**زبان‌ها**  
- فارسی (زبان مادری)  
- انگلیسی (متوسط، توانایی مطالعه و نوشتن مستندات فنی)

**اهداف حرفه‌ای**  
- توسعه دانش و مهارت در زمینه یادگیری ماشین و هوش مصنوعی  
- تمرکز بر پروژه‌های تحلیل داده و پیاده‌سازی راه‌حل‌های پیشرفته برای مدیریت و بهبود سیستم‌های پایگاه داده
